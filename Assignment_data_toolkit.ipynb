Question-1. What is NumPy, and why is it widely used in Python?

            NumPy is a Python library used for numerical and scientific computing.
            It is widely used because:

            It supports fast array operations

            It provides multidimensional arrays

            It enables vectorized operations

            It integrates well with other libraries like Pandas, SciPy, Scikit-learn.

Question-2. How does broadcasting work in NumPy?

            Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding them to compatible shapes.
            Example: Adding a scalar to an array → scalar is broadcast across all elements.

Question-3. What is a Pandas DataFrame?

            A DataFrame is a 2D tabular data structure in Pandas with rows and columns, similar to an Excel sheet or SQL table.

Question-4. Explain the use of groupby() in Pandas.

            groupby() groups rows based on column values and allows aggregated operations like sum, mean, min, max, count.

Question-5. Why is Seaborn preferred for statistical visualizations?

            Because it:

            Provides beautiful, high-level statistical plots

            Works on top of Matplotlib

            Automatically handles themes, color palettes, confidence intervals

            Integrates easily with Pandas DataFrames

Question-6. Differences between NumPy arrays and Python lists

            NumPy Array	Python List
            Fast & memory efficient	Slower
            Stores same data type	Can store mixed types
            Supports vectorized ops	Requires loops
            Designed for math operations	Not optimized for math
